```python
import pandas as pd
```

```python
incidencia_tb = pd.read_csv('TB_burden_countries_2019-05-23.csv')
incidencia_tb
```

```python
incidencias = incidencia_tb.groupby('iso2')['e_inc_num'].apply(list)
incidencias
```

```python
from sklearn import linear_model
import numpy as np
import matplotlib.pyplot as plt
for l in incidencias:
    n_samples, n_features = len(l), 1
    x = np.array(range(0,n_samples)).reshape(-1, 1)
    y = np.array(l)
    regr = linear_model.LinearRegression()
    regr.fit(x, y)
    a = regr.predict(np.array([len(l)]).reshape(-1,1))[0]
    if a < 0:
        a = 0
    l.append(a)
```
