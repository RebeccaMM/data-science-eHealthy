```python
incidencia_tb = pd.read_csv('TB_burden_countries_2019-05-23.csv')
incidencia_tb
```

```python
incidencia2017 = incidencia_tb.loc[incidencia_tb['year'] == 2017][['iso2', 'e_inc_num']]
incidencia2017.columns = ['id', 'e_inc_num']
incidencia2017.to_csv('incidencia2017.csv', index=False)
incidencia2017
```

```python
import numpy as np
import ast

df = pd.read_csv('fronteiras.csv')
x = df.values
adjacent = []
for c in x:
    if c[0] in incidencia2017['iso2'].values:
        for c2 in ast.literal_eval(c[1]):
            adjacent.append(c[0] + ',' + c2 + '\n')
```

```python
with open("adjacencia.csv", "w") as f:
    for line in adjacent:
        f.write('source,target\n')
        f.writelines(adjacent)
```

```python
idade_df = pd.read_csv('TB_burden_age_sex_2019-05-23.csv')
idade_df
```

```python
total_f = np.sum(idade_df.loc[idade_df['sex'] == 'f']['best'].values)
total_m = np.sum(idade_df.loc[idade_df['sex'] == 'm']['best'].values)
```

```python
import matplotlib.pyplot as plt

# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = 'Feminino', 'Masculino'
total = total_f + total_m
sizes = [total_f, total_m]

fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

```python
age_groups = idade_df['age_group'].unique()
age_groups = np.delete(age_groups, np.where(age_groups == 'all'))
age_groups = np.delete(age_groups, np.where(age_groups == '15plus'))
age_sums = []
for group in age_groups:
    age_sums.append(np.sum(idade_df.loc[idade_df['age_group'] == group]['best'].values))

# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = age_groups
sizes = age_sums

fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```
